```python
#There are total 168 entries in customers df and just 161 unique email addresses, there must repetitions.
#Isolate repeated customer emails.

emails=[]
duplicates = 0
frequency_emails={}
for i in range(len(customers_df.index)):
    duplicates= 0
    for j in range(len(customers_df.index)):
        if customers_df['customerEmail'][i] == customers_df['customerEmail'][j]:
            duplicates+=1
    frequency_emails.update({customers_df['customerEmail'][i]:duplicates})
frequency_emails    

duplicates = []

for (emails, repeates) in frequency_emails.items():
    if (repeates > 1):
        duplicates.append(emails)         
duplicates
```

```python
#Isolate all fradulent customer emails and check if 
fradulent_customers = customers_df[customers_df['Fraud'] == True]

#Check if our duplicated email is among fradulent customers

fradulent_customers[fradulent_customers['customerEmail'] == 'johnlowery@gmail.com']

```

#### Changing the column names to prepare for our join of the two datasets 


```python
# Used the merge() method to join Customers to transactions database and drop duplicates
df = pd.merge(transactions_df, customers_df, on='customerEmail')

df = df.drop_duplicates(subset='transactionId')
```

```python
# Check for duplicates in joint database
df_duplicates = df.duplicated().sum()

df_duplicates
```

```python
# Check for missing values in joint database
missing_values = df.customerEmail.isnull().sum()

missing_values
```

```python
# Separate states for fraud cases

df["State"] = df["customerBillingAddress"].str.split(",", expand=True)[1]

df["State"] = df['State'].str.replace('\d+', '')

df["State"] = df['State'].str.replace(r"[^a-zA-Z0-9]", '')

df
```

```python
# save the data to a new csv file
datapath = '../data/summary_data_cleaned.csv'

df.to_csv(datapath)
```
